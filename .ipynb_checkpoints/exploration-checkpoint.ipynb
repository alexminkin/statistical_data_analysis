## Data preprocessing

Convert features with dates to the correct format

````python
users['reg_date'] = pd.to_datetime(users['reg_date'], format='%Y-%m-%d')
users['churn_date'] = pd.to_datetime(users['churn_date'], format='%Y-%m-%d')
calls['call_date'] = pd.to_datetime(calls['call_date'], format='%Y-%m-%d')
messages['message_date'] = pd.to_datetime(messages['message_date'], format='%Y-%m-%d')
sessions['session_date'] = pd.to_datetime(sessions['session_date'], format='%Y-%m-%d')
````

In feature "duration", round the values up and change the data type

````python
calls['duration'] = np.ceil(calls['duration'])
calls['duration'] = calls['duration'].astype('int')
````

Delete feature 'Unnamed: 0'

````python
sessions = sessions.drop(['Unnamed: 0'], axis=1)
````

Create feature 'month'

````python
calls['month'] = calls['call_date'].dt.month
messages['month'] = messages['message_date'].dt.month
sessions['month'] = sessions['session_date'].dt.month
````

Let's group the data by each user and month

````python
# counting the number of calls for each user by month
calls_per_month = calls.groupby(by=['user_id', 'month']).agg(calls=('duration', 'count'))
# calculation of spent minutes for each user by month
minutes_per_month = calls.groupby(by=['user_id', 'month']).agg(minutes=('duration', 'sum'))
# counting the number of sent messages for each user by month
messages_per_month = messages.groupby(by=['user_id', 'month']).agg(messages=('message_date', 'count'))
# calculation of spent megabytes for each user by months
sessions_per_month = sessions.groupby(by=['user_id', 'month']).agg({'mb_used': 'sum'})
````

## Data analysis

Find out the share of terminated contracts with customer

````python
users['churn_date'].count() / users['churn_date'].shape[0] * 100
````

Create new dataset, which will contain **calls_per_month, minutes_per_month, messages_per_month, sessions_per_month**

````python
user_behavior = calls_per_month\
    .merge(messages_per_month, left_index=True, right_index=True, how='outer')\
    .merge(sessions_per_month, left_index=True, right_index=True, how='outer')\
    .merge(minutes_per_month, left_index=True, right_index=True, how='outer')\
    .reset_index()\
    .merge(users, how='left', left_on='user_id', right_on='user_id')\

user_behavior.head()
````

Let's check missing values in user_behavior

````python
user_behavior.isna().sum()
````

Let's fill in missing values in the **user_behavior**

````python
user_behavior['calls'] = user_behavior['calls'].fillna(0)
user_behavior['minutes'] = user_behavior['minutes'].fillna(0)
user_behavior['messages'] = user_behavior['messages'].fillna(0)
user_behavior['mb_used'] = user_behavior['mb_used'].fillna(0)
````

Let's add feature **tariffs**

````python
tariffs = tariffs.rename(
    columns={
        'tariff_name': 'tariff'
    }
)

user_behavior = user_behavior.merge(tariffs, on='tariff')
````

Let's calculate minutes, messages, traffic amount 

````python
user_behavior['paid_minutes'] = user_behavior['minutes'] - user_behavior['minutes_included']
user_behavior['paid_messages'] = user_behavior['messages'] - user_behavior['messages_included']
user_behavior['paid_mb'] = user_behavior['mb_used'] - user_behavior['mb_per_month_included']

for col in ['paid_messages', 'paid_minutes', 'paid_mb']:
    user_behavior.loc[user_behavior[col] < 0, col] = 0
    
````

Let's make new feature **paid_gb**

````python
user_behavior['paid_gb'] = np.ceil(user_behavior['paid_mb'] / 1024).astype(int)
````

Let's make new feature, which is business metrics

````python
user_behavior['cost_minutes'] = user_behavior['paid_minutes'] * user_behavior['rub_per_minute']
user_behavior['cost_messages'] = user_behavior['paid_messages'] * user_behavior['rub_per_message']
user_behavior['cost_gb'] = user_behavior['paid_gb'] * user_behavior['rub_per_gb']
````

````python
user_behavior['total_cost'] = \
      user_behavior['rub_monthly_fee']\
    + user_behavior['cost_minutes']\
    + user_behavior['cost_messages']\
    + user_behavior['cost_gb']
````

Let's make new DataFrame **stats_df**, which contains main metrics. This DataFrame we will use for data analysis

````python
stats_df = user_behavior.pivot_table(
            index=['month', 'tariff'],\
            values=['calls', 'minutes', 'messages', 'mb_used'],\
            aggfunc=['mean', 'std', 'median']\
).round(2).reset_index()

stats_df.columns=['month', 'tariff', 'calls_mean', 'sessions_mean', 'messages_mean', 'minutes_mean',
                                     'calls_std',  'sessions_std', 'messages_std', 'minutes_std', 
                                     'calls_median', 'sessions_median', 'messages_median',  'minutes_median']

stats_df.head(10)

````

## Test hypotheses

Test hypotheses: The average revenue of Smart and Ultra tariffs differs

```
H_0: total_cost Ultra users = total_cost Smart users`
H_1: total_cost Ultra users ≠ total_cost Smart users`
alpha = 0.05
```

````python
total_smart = user_behavior.loc[user_behavior['tariff'] == 'smart', 'total_cost']
total_ultra = user_behavior.loc[user_behavior['tariff'] == 'ultra', 'total_cost']

results = st.ttest_ind(total_smart, total_ultra, equal_var = False)

alpha = .05

print(results.pvalue)

if results.pvalue < alpha:
    print('Reject H_0')
else:
    print('Do not reject H_0')
````

Test hypotheses: Moscow users bring in more revenue, than other city users

```
H_0: total_cost Moscow users = total_cost other cities users`
H_1: total_cost Moscow users ≠ total_cost other cities users`
alpha = 0.05
```

````python
total_moscow = user_behavior.loc[user_behavior['city'] == 'Москва', 'total_cost']
total_not_moscow = user_behavior.loc[user_behavior['city'] != 'Москва', 'total_cost']

results = st.ttest_ind(total_moscow, total_not_moscow, equal_var = False)

alpha = .05

print(results.pvalue)

if results.pvalue < alpha:
    print('Reject H_0')
else:
    print('Do not reject H_0')
````

## Let's prepare data for training

Let's choose necessary features

````python
df = pd.DataFrame()
df = user_behavior[['calls', 'minutes', 'messages', 'mb_used', 'tariff']]
df = df.rename(columns={'tariff': 'is_ultra'})
df.loc[df['is_ultra'] == 'ultra', 'is_ultra'] = '1'
df.loc[df['is_ultra'] == 'smart', 'is_ultra'] = '0'
df['is_ultra'] = df['is_ultra'].astype('int32')
df.info()
````

````python
df.head()
````

Let's explore data for multicollinearity

````python
sns.pairplot(df, hue='is_ultra')
````

````python
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12)
````

Remove feature 'calls'

````python
df = df.drop(['calls'], axis=1)
df.head()
````

Let's split the dataset into train, valid and test samples

````python
features = df.drop(['is_ultra'], axis=1)
target = df['is_ultra']
````

````python
features_train_valid, features_test, target_train_valid, target_test = train_test_split(features, target, test_size=0.2, random_state=12345)
features_train, features_valid, target_train, target_valid = train_test_split(features_train_valid, target_train_valid, test_size=0.25, random_state=12345)

print(features_train.shape)
print(features_valid.shape)
print(features_test.shape)
````

## Train models

**DecisionTreeClassifier**

````python
best_model_tree = None
best_result = 0
best_depth = 0
for depth in range(1, 21):
    model_tree = DecisionTreeClassifier(random_state=12345, max_depth=depth)
    model_tree.fit(features_train, target_train)

    predictions = model_tree.predict(features_valid)
    result = accuracy_score(target_valid, predictions)
    if result > best_result:
        best_result = result
        best_model_tree = model_tree
        best_depth = depth
print('Best accuracy:', best_result)
print('depth =', best_depth)
````

````python
best_model_tree = None
best_result = 0
best_depth = 0
for leaf in range(1, 25):
    for depth in range(1, 21):
        model_tree = DecisionTreeClassifier(random_state=12345, max_depth=depth, min_samples_leaf=leaf)
        model_tree.fit(features_train, target_train)

        predictions = model_tree.predict(features_valid)
        result = accuracy_score(target_valid, predictions)
        if result > best_result:
            best_result = result
            best_model_tree = model_tree
            best_depth = depth
print('Best accuracy:', best_result)
print('depth =', best_depth)
````

````python
best_model_tree = None
best_result = 0
best_depth = 0
best_leaf = 0
for leaf in range(1, 25):
    for depth in range(1, 21):
        model_tree = DecisionTreeClassifier(random_state=12345, max_depth=depth, min_samples_leaf=leaf, splitter='random')
        model_tree.fit(features_train, target_train)

        predictions = model_tree.predict(features_valid)
        result = accuracy_score(target_valid, predictions)
        if result > best_result:
            best_result = result
            best_model_tree = model_tree
            best_depth = depth
            best_leaf = leaf
print('Best accuracy:', best_result)
print('depth =', best_depth)
print('leaf =', best_leaf)
````

The best result is achieved with the following hyperparameters:
- depth = 9
- leaf = 9
- splitter = random

**RandomForestClassifier**

````python
best_model_forest_1 = None
best_result = 0
best_est = 0
best_depth = 0

for est in range(10, 201, 10):
    for depth in range(1, 16):
        model_forest = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth)
        model_forest.fit(features_train, target_train)
        
        predictions = model_forest.predict(features_valid)
        result = accuracy_score(target_valid, predictions)
        if result > best_result:
            best_result = result
            best_model_forest_1 = model_forest
            best_est = est
            best_depth = depth
print('Best accuracy', best_result)
print('Best est', best_est)
print('Best depth', best_depth)
````

````python
best_model_forest_3 = None
best_result = 0
best_est = 0
best_depth = 0
best_leaf = 0

for leaf in range(1, 25):
    for est in range(10, 201, 10):
        for depth in range(1, 16):
            model_forest = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth, min_samples_leaf=leaf)
            model_forest.fit(features_train, target_train)

            predictions = model_forest.predict(features_valid)
            result = accuracy_score(target_valid, predictions)
            if result > best_result:
                best_result = result
                best_model_forest_3 = model_forest
                best_est = est
                best_depth = depth
                best_leaf = leaf
print('Best accuracy', best_result)
print('Best est', best_est)
print('Best depth', best_depth)
print('Best leaf', best_leaf)
````

````python
THRESHOLD = 0.01
best_model_forest_2 = None
best_crit = None
best_result = 0
best_est = 0
best_depth = 0
criterions = ['gini', 'entropy', 'log_loss']

list_accuracy = []

for crit in criterions:
    for est in range(10, 201, 10):
        for depth in range(1, 16):
            model_forest = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth, criterion=crit)
            model_forest.fit(features_train, target_train)

            predictions = model_forest.predict(features_valid)
            result = accuracy_score(target_valid, predictions)
            list_accuracy.append(result)
            if len(list_accuracy) == 10:
                std = np.std(list_accuracy)
                if std <= THRESHOLD:
                    break
            elif len(list_accuracy) > 10:
                list_accuracy.pop(0)
                
            if result > best_result:
                best_result = result
                best_model_forest_2 = model_forest
                best_est = est
                best_depth = depth
                best_crit = crit
        if std <= THRESHOLD:
            break
    if std <= THRESHOLD:
        break
print('Best accuracy', best_result)
print('Best est', best_est)
print('Best depth', best_depth)
print('Best criterion', best_crit)
````

The best result is achieved with the following hyperparameters:
- Best est = 50
- Best depth = 15
- Best leaf = 8
- criterion = entropy

**LogisticRegression**

````python
best_model_log = None
best_solver = None
best_result = 0
best_iter = 0

solvers = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']

for sol in solvers:
    for iters in range(500, 5000, 500):
        model_log = LogisticRegression(random_state=12345, solver=sol, max_iter=iters)
        model_log.fit(features_train, target_train)
        predictions = model_log.predict(features_valid)
        result = accuracy_score(target_valid, predictions)
        if result > best_result:
            best_result = result
            best_model_log = model_log
            best_iter = iters
            best_solver = sol
print('Best accuracy', best_result)
print('Best iter', best_iter)
print('Best solver', best_solver)
````

**Conclusion:** The best model is RandomForestClassifier with accuracy = 0.8211
